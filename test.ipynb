```python
E = torch.eye(16)
P = E[torch.randperm(16)]

plt.imshow(P)
```

```python
a = torch.rand(1, 16, 3)
v = P @ a

plt.imshow(a[[0]].detach().cpu())
plt.show()

plt.imshow(v[[0]].detach().cpu())
plt.show()
```

```python
at = a.transpose(1, 2)

ai = torch.inverse(a @ at)
i = at @ ai

plt.imshow(i[[0]].transpose(1, 2))
plt.show()
```

```python
PP = v @ i

plt.imshow(PP[0])
plt.show()
```
